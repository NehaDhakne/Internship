# Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release)

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
```

```python
soup = BeautifulSoup(page.content)
```

```python
title = soup.find_all('td', class_="titleColumn")
title
```

```python
titles = []

for i in title:
    titles.append(i.text.replace('\n',''))

titles
```

```python
rating = soup.find_all('td', "ratingColumn imdbRating")
```

```python
rating
```

```python
ratings = []

for i in rating:
    ratings.append(i.text.strip())

ratings
```

```python
import pandas as pd

movies = pd.DataFrame({})

movies['Movies with year of release'] = titles
movies['IMDB Ratings'] = ratings

movies
```
