```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
```

```python
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
# lets extract team names

teams=soup.find_all('span', class_="u-hide-phablet")
```

```python
teams
```

```python
teams_1=[]

for i in teams:
    teams_1.append(i.text)
    
teams_1
```

```python
final_teams = teams_1[0:10]
```

```python
final_teams
```
