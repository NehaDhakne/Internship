Question4: Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
from bs4 import BeautifulSoup
import requests
```

Our table should have following columns

1) Book 

2) Author 

3) Genre

4) Book review

```python
page = requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content)
```

```python
soup
```

```python
print(soup.prettify())
```

```python
# extract firstbook title

first_book = soup.find('h4', class_='italic')

first_book
```

```python
first_book.text.strip()
```

```python
# lets extract all the book titles

book_names= soup.find_all('h4', class_='italic')

book_names
```

```python
# we have all the tags with book names
# we have to extract these tags one by one, for which we will use loops


book_titles = []  
for i in book_names:
    book_titles.append(i.text.strip())
book_titles
```

```python
# we have to extract the first author of the first book

first_author=soup.find('p', class_='sans bold')

first_author
```

```python
first_author.text.strip()
```

```python
# lets extract all the authors

author_names = soup.find_all('p', class_='sans bold')

author_names
```

```python
#we have all the tags with author names
# we have to extract the names of authors one by one, we will use the for loops

author_titles = []
for i in author_names:
    author_titles.append(i.text.strip())
author_titles
```

```python
#Let's extract the first book's genre

first_genre = soup.find('p', class_='genre-links hidden-phone')

first_genre
```

```python
first_genre.text
```

```python
first_genre.text.replace('\n','')
```

```python
# lets extract all the genres

genre_names = soup.find_all('p', class_='genre-links hidden-phone')

genre_names
```

```python

# we have all the tags with genres
# we will use for loop

genre_titles = []
for i in genre_names:
    genre_titles.append(i.text.replace("\n", ''))
genre_titles
```

```python
# lets extract the first book review

first_review = soup.find('p', class_="excerpt")
first_review
```

```python
first_review.text.strip()
```

```python
#lets extract all the genres in relation to books

book_review = soup.find_all('p', class_='excerpt')

book_review
```

```python
# using for loop

reviews = []

for i in book_review:
    reviews.append(i.text.strip())

reviews
```

```python
import pandas as pd

bookss = pd.DataFrame({})

bookss['Names']=book_titles
bookss['Authors']=author_titles
bookss['Genres']=genre_titles
bookss['Reviews']=reviews
```
